```python
import os
import pandas as pd
from collections import defaultdict
from tqdm import tqdm

# === Укажи путь до папки с окнами ===
BASE_FOLDER = r"C:\ML-itcamp\split_windows_2048"  # Подставь свой путь

# === Статистика: общее и аномальные окна по фазам ===
stats = defaultdict(lambda: {"total": 0, "empty_phase_windows": 0})

# === Собираем только числовые подпапки (например: '1', '2', ..., '38') ===
dataset_ids = [x for x in os.listdir(BASE_FOLDER) if x.isdigit()]

# === Сначала посчитаем общее количество окон ===
all_window_paths = []
for dataset_id in dataset_ids:
    dataset_path = os.path.join(BASE_FOLDER, dataset_id)
    if not os.path.isdir(dataset_path):
        continue
    for file in os.listdir(dataset_path):
        if file.endswith(".csv"):
            all_window_paths.append((dataset_id, os.path.join(dataset_path, file)))

print(f"🔢 Всего найдено окон: {len(all_window_paths)}")

# === Анализ с отображением прогресса ===
for dataset_id, file_path in tqdm(all_window_paths, desc="Проверка окон", unit="окно"):
    stats[dataset_id]["total"] += 1

    try:
        df = pd.read_csv(file_path)

        # Проверка: хотя бы одна из фаз пустая или постоянная
        if (
            "current_R" not in df.columns or df["current_R"].nunique() <= 1 or
            "current_S" not in df.columns or df["current_S"].nunique() <= 1 or
            "current_T" not in df.columns or df["current_T"].nunique() <= 1
        ):
            stats[dataset_id]["empty_phase_windows"] += 1

    except Exception:
        stats[dataset_id]["empty_phase_windows"] += 1

# === Формируем финальную таблицу ===
df_result = pd.DataFrame([
    {
        "dataset_id": dataset,
        "total_windows": values["total"],
        "empty_phase_windows": values["empty_phase_windows"]
    }
    for dataset, values in stats.items()
])

df_result = df_result.sort_values(by="empty_phase_windows", ascending=False).reset_index(drop=True)
df_result
```
