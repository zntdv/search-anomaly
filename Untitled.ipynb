```python
# 📌 1. Импорты и настройки
import pandas as pd
import matplotlib.pyplot as plt
import os
from IPython.display import display
import ipywidgets as widgets

# Папки и файлы
WINDOWS_DIR = "split_windows"
LABELS_FILE = "labels.csv"

# Метки
label_dict = {
    0: "0 — нет дефекта",
    1: "1 — дефект подшипника",
    2: "2 — дисбаланс",
    3: "3 — расцентровка",
    4: "4 — прочее"
}
label_options = list(label_dict.items())

# Загрузка уже размеченных
if os.path.exists(LABELS_FILE):
    labels_df = pd.read_csv(LABELS_FILE)
    labeled_files = set(labels_df['filename'])
else:
    labels_df = pd.DataFrame(columns=["filename", "label"])
    labeled_files = set()

# Список файлов
window_files = sorted([f for f in os.listdir(WINDOWS_DIR) if f.endswith(".csv")])
unlabeled_files = [f for f in window_files if f not in labeled_files]
print(f"🗃️ Осталось размечать: {len(unlabeled_files)} окон")

# 📌 2. Выбор файла
file_dropdown = widgets.Dropdown(
    options=unlabeled_files,
    description='Файл:',
    layout=widgets.Layout(width='400px')
)
display(file_dropdown)

# 📌 3. Визуализация сигнала
def plot_window(file_name):
    df = pd.read_csv(os.path.join(WINDOWS_DIR, file_name))
    plt.figure(figsize=(14, 4))
    plt.plot(df['current_R'], label='Фаза R')
    plt.plot(df['current_S'], label='Фаза S')
    plt.plot(df['current_T'], label='Фаза T')
    plt.title(f"Сигналы тока — {file_name}")
    plt.xlabel("Индекс")
    plt.ylabel("Ток (А)")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

plot_window(file_dropdown.value)

# 📌 4. Разметка и сохранение
label_dropdown = widgets.Dropdown(
    options=[(v, k) for k, v in label_dict.items()],  # (label_text, value)
    description='Метка:',
    layout=widgets.Layout(width='300px')
)

save_button = widgets.Button(description="💾 Сохранить метку", button_style="success")

output = widgets.Output()

def on_save_clicked(b):
    filename = file_dropdown.value
    label = label_dropdown.value
    global labels_df
    labels_df = pd.concat([labels_df, pd.DataFrame([{"filename": filename, "label": label}])], ignore_index=True)
    labels_df.to_csv(LABELS_FILE, index=False)
    with output:
        output.clear_output()
        print(f"✅ Метка для {filename} сохранена как {label_dict[label]}")
        unlabeled_files.remove(filename)
        if unlabeled_files:
            file_dropdown.options = unlabeled_files
            file_dropdown.value = unlabeled_files[0]
            plot_window(file_dropdown.value)
        else:
            print("🎉 Все окна размечены!")

save_button.on_click(on_save_clicked)

display(label_dropdown, save_button, output)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import ipywidgets as widgets
from IPython.display import display

# === Параметры ===
WINDOWS_DIR = "split_windows_1"

# Получение списка CSV-файлов
window_files = sorted([f for f in os.listdir(WINDOWS_DIR) if f.endswith(".csv")])

# Выпадающий список файлов
file_dropdown = widgets.Dropdown(
    options=window_files,
    description="Файл:",
    layout=widgets.Layout(width='400px')
)

# Кнопка отрисовки
plot_button = widgets.Button(description="📈 Показать график", button_style="info")
output = widgets.Output()

# Функция отрисовки
def plot_selected_file(b):
    output.clear_output()
    with output:
        selected_file = file_dropdown.value
        df = pd.read_csv(os.path.join(WINDOWS_DIR, selected_file))
        
        plt.figure(figsize=(12, 4))
        plt.plot(df["current_R"], label="Фаза R")
        plt.plot(df["current_S"], label="Фаза S")
        plt.plot(df["current_T"], label="Фаза T")
        plt.title(f"Токовые сигналы — {selected_file}")
        plt.xlabel("Индекс")
        plt.ylabel("Ток (А)")
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()

plot_button.on_click(plot_selected_file)

# Отображаем виджеты
display(file_dropdown, plot_button, output)
```

```python
import os
import pandas as pd
import numpy as np

# === Настройки ===
BASE_DIR = "split_windows"  # Папка с датасетами по номерам
records = []

# Получаем список всех подкаталогов с цифрами (0, 1, 2, ..., 37)
dataset_folders = sorted([f for f in os.listdir(BASE_DIR) if f.isdigit()], key=lambda x: int(x))

for dataset_id in dataset_folders:
    print(f"🔍 Обрабатывается датасет {dataset_id}...")
    dataset_path = os.path.join(BASE_DIR, dataset_id)
    if not os.path.isdir(dataset_path):
        continue

    for file in sorted(os.listdir(dataset_path)):
        if not file.endswith(".csv"):
            continue
        try:
            file_path = os.path.join(dataset_path, file)
            df = pd.read_csv(file_path)

            if not {"current_R", "current_S", "current_T"}.issubset(df.columns):
                print(f"⚠️ Пропущен файл {file_path} — отсутствуют колонки.")
                continue

            stds = df.std().to_dict()
            energy = np.sum(df.values**2)
            phase_diff = (
                np.mean(np.abs(df["current_R"] - df["current_S"])) +
                np.mean(np.abs(df["current_S"] - df["current_T"])) +
                np.mean(np.abs(df["current_T"] - df["current_R"]))
            )

            records.append({
                "dataset_id": int(dataset_id),
                "window": file,
                "std_mean": np.mean([stds["current_R"], stds["current_S"], stds["current_T"]]),
                "energy": energy,
                "phase_diff": phase_diff
            })

        except Exception as e:
            print(f"❌ Ошибка в файле {file_path}: {e}")

# === Формирование DataFrame и очистка от NaN ===
df_windows = pd.DataFrame(records)
df_windows = df_windows.dropna(subset=["dataset_id", "std_mean", "energy", "phase_diff"])

# === Вычисляем TOP-1 по каждой метрике ===
top_std = df_windows.loc[df_windows.groupby("dataset_id")["std_mean"].idxmax()].sort_values("std_mean", ascending=False)
top_energy = df_windows.loc[df_windows.groupby("dataset_id")["energy"].idxmax()].sort_values("energy", ascending=False)
top_phase = df_windows.loc[df_windows.groupby("dataset_id")["phase_diff"].idxmax()].sort_values("phase_diff", ascending=False)

# === Вывод результатов ===
print("🏆 Топ-1 по std_mean:")
display(top_std.head(1))

print("🏆 Топ-1 по energy:")
display(top_energy.head(1))

print("🏆 Топ-1 по phase_diff:")
display(top_phase.head(1))
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output

# === Папка с окнами ===
BASE_DIR = "split_windows_256"

# === Получаем список доступных датасетов ===
available_datasets = sorted([f for f in os.listdir(BASE_DIR) if f.isdigit()], key=lambda x: int(x))
dataset_dropdown = widgets.Dropdown(
    options=available_datasets,
    description="Датасет:",
    layout=widgets.Layout(width="300px")
)

# === Ввод списка окон ===
windows_input = widgets.Text(
    value="0, 1, 2",
    description="Окна:",
    layout=widgets.Layout(width="300px"),
    placeholder="например: 0, 2, 5"
)

# === Кнопка и вывод графиков ===
plot_button = widgets.Button(description="📈 Показать окна", button_style="info")
output_plot = widgets.Output()

# === Построение графиков ===
def show_multiple_plots(b=None):
    dataset_id = dataset_dropdown.value
    window_ids = windows_input.value.replace(" ", "").split(",")

    with output_plot:
        clear_output(wait=True)

        for window_id in window_ids:
            try:
                window_file = f"window_{int(window_id):03d}.csv"
                file_path = os.path.join(BASE_DIR, dataset_id, window_file)
                df = pd.read_csv(file_path)

                plt.figure(figsize=(12, 3))
                plt.plot(df["current_R"], label="Фаза R")
                plt.plot(df["current_S"], label="Фаза S")
                plt.plot(df["current_T"], label="Фаза T")
                plt.title(f"Датасет {dataset_id} — {window_file}")
                plt.xlabel("Индекс")
                plt.ylabel("Ток (А)")
                plt.grid(True)
                plt.legend()
                plt.tight_layout()
                plt.show()

            except Exception as e:
                print(f"⚠️ Ошибка при чтении окна {window_id}: {e}")

# === Привязка кнопки ===
plot_button.on_click(show_multiple_plots)

# === Интерфейс ===
display(widgets.HBox([dataset_dropdown, windows_input]), plot_button, output_plot)
```
