```python

# 📓 Анализ аномалий: топ-5 окон по каждой метрике

import os
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis
from scipy.signal import find_peaks
from scipy.fft import rfft, rfftfreq

# === Параметры ===
BASE_DIR = "split_windows_256"  # Укажи папку с окнами
WINDOW_SIZE = 256

records = []

# Получаем список датасетов
dataset_folders = sorted([f for f in os.listdir(BASE_DIR) if f.isdigit()], key=lambda x: int(x))

for dataset_id in dataset_folders:
    dataset_path = os.path.join(BASE_DIR, dataset_id)
    if not os.path.isdir(dataset_path):
        continue

    for file in sorted(os.listdir(dataset_path)):
        if not file.endswith(".csv"):
            continue
        try:
            file_path = os.path.join(dataset_path, file)
            df = pd.read_csv(file_path)

            data = df[["current_R", "current_S", "current_T"]].to_numpy()
            avg_signal = data.mean(axis=1)

            std_mean = np.mean(np.std(data, axis=0))
            energy = np.sum(avg_signal ** 2)
            phase_diff = (
                np.mean(np.abs(data[:, 0] - data[:, 1])) +
                np.mean(np.abs(data[:, 1] - data[:, 2])) +
                np.mean(np.abs(data[:, 2] - data[:, 0]))
            )

            skewness = skew(avg_signal)
            kurt = kurtosis(avg_signal)
            zcr = np.mean(np.diff(np.sign(avg_signal)) != 0)
            peaks, _ = find_peaks(avg_signal)
            peak_count = len(peaks)

            spectrum = np.abs(rfft(avg_signal))
            freqs = rfftfreq(len(avg_signal), d=1)
            fft_max_amp = np.max(spectrum)
            dominant_freq = freqs[np.argmax(spectrum)]

            records.append({
                "dataset_id": int(dataset_id),
                "window_file": file,
                "std_mean": std_mean,
                "energy": energy,
                "phase_diff": phase_diff,
                "skewness": skewness,
                "kurtosis": kurt,
                "zcr": zcr,
                "peak_count": peak_count,
                "fft_max_amp": fft_max_amp,
                "dominant_freq": dominant_freq
            })

        except Exception as e:
            print(f"❌ Ошибка в файле {file_path}: {e}")

# Формируем общий DataFrame
df = pd.DataFrame(records)
df.head()
```

```python

# === Топ-5 самых аномальных окон по каждой метрике ===

metrics = [
    "std_mean", "energy", "phase_diff",
    "skewness", "kurtosis", "zcr",
    "peak_count", "fft_max_amp", "dominant_freq"
]

top_by_metric = {}
for metric in metrics:
    top_by_metric[metric] = df.sort_values(by=metric, ascending=False).head(5)

# Отображение результатов
for metric in metrics:
    print(f"\n📊 Топ-5 по метрике: {metric}")
    display(top_by_metric[metric])
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output

# === Папка с окнами ===
BASE_DIR = "split_windows_256"

# === Получаем список доступных датасетов ===
available_datasets = sorted([f for f in os.listdir(BASE_DIR) if f.isdigit()], key=lambda x: int(x))
dataset_dropdown = widgets.Dropdown(
    options=available_datasets,
    description="Датасет:",
    layout=widgets.Layout(width="300px")
)

# === Ввод списка окон ===
windows_input = widgets.Text(
    value="0, 1, 2",
    description="Окна:",
    layout=widgets.Layout(width="300px"),
    placeholder="например: 0, 2, 5"
)

# === Кнопка и вывод графиков ===
plot_button = widgets.Button(description="📈 Показать окна", button_style="info")
output_plot = widgets.Output()

# === Построение графиков ===
def show_multiple_plots(b=None):
    dataset_id = dataset_dropdown.value
    window_ids = windows_input.value.replace(" ", "").split(",")

    with output_plot:
        clear_output(wait=True)

        for window_id in window_ids:
            try:
                window_file = f"window_{int(window_id):03d}.csv"
                file_path = os.path.join(BASE_DIR, dataset_id, window_file)
                df = pd.read_csv(file_path)

                plt.figure(figsize=(12, 3))
                plt.plot(df["current_R"], label="Фаза R")
                plt.plot(df["current_S"], label="Фаза S")
                plt.plot(df["current_T"], label="Фаза T")
                plt.title(f"Датасет {dataset_id} — {window_file}")
                plt.xlabel("Индекс")
                plt.ylabel("Ток (А)")
                plt.grid(True)
                plt.legend()
                plt.tight_layout()
                plt.show()

            except Exception as e:
                print(f"⚠️ Ошибка при чтении окна {window_id}: {e}")

# === Привязка кнопки ===
plot_button.on_click(show_multiple_plots)

# === Интерфейс ===
display(widgets.HBox([dataset_dropdown, windows_input]), plot_button, output_plot)
```
