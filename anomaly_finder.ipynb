```python

# 📓 Auto-anomaly Finder: поиск "странных" токовых сигналов среди окон

import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt

# === Параметры ===
WINDOWS_DIR = "split_windows"
MAX_WINDOWS = 1000  # ограничим число окон для ускорения, если надо

# === Сбор фичей для всех окон ===
features = []

files = sorted([f for f in os.listdir(WINDOWS_DIR) if f.endswith(".csv")])[:MAX_WINDOWS]

for file in files:
    path = os.path.join(WINDOWS_DIR, file)
    df = pd.read_csv(path)

    stds = df.std().values
    means = df.mean().values
    energy = np.sum(df.values**2)
    phase_diff = (
        np.mean(np.abs(df['current_R'] - df['current_S'])) +
        np.mean(np.abs(df['current_S'] - df['current_T'])) +
        np.mean(np.abs(df['current_T'] - df['current_R']))
    )

    features.append({
        "filename": file,
        "std_R": stds[0],
        "std_S": stds[1],
        "std_T": stds[2],
        "mean_R": means[0],
        "mean_S": means[1],
        "mean_T": means[2],
        "phase_diff": phase_diff,
        "energy": energy
    })

df_feat = pd.DataFrame(features)

# === Сортировка по "подозрительности" ===
df_sorted = df_feat.sort_values(by="phase_diff", ascending=False)
df_sorted.head(30)
```
